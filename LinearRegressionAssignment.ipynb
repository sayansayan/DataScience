```python
from sklearn.model_selection import train_test_split
```

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
day_train, day_test = train_test_split(day, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['temp', 'hum', 'windspeed']

day_train[num_vars] = scaler.fit_transform(day_train[num_vars])
```

```python
y_train = day_train.pop('cnt')
X_train = day_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
print(lm.summary())
```

```python
X_train_new = X_train_rfe.drop(['mnth'], axis = 1)
X_train_new
```

```python
# Rebuilding the model 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()
```

```python
print(lm.summary())
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train_new = X_train_new.drop(['workingday'], axis = 1)
X_train_new
```

```python
X_train_lm = sm.add_constant(X_train_new)
lm = sm.OLS(y_train,X_train_lm).fit()
print(lm.summary())
```

```python
X_train_new.columns
```

```python
X_train_new = X_train_new.drop(['hum'], axis = 1)
X_train_new
X_train_lm = sm.add_constant(X_train_new)
lm = sm.OLS(y_train,X_train_lm).fit()
print(lm.summary())
```

```python
X_train_new = X_train_new.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
y_train_cnt = lm.predict(X_train_lm)
```

```python
# Importing the required libraries for plots.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18) 
```

```python
num_vars = ['temp', 'hum', 'windspeed']

day_test[num_vars] = scaler.fit_transform(day_test[num_vars])
```

```python
y_test = day_test.pop('cnt')
X_test = day_test
```

```python
# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```
